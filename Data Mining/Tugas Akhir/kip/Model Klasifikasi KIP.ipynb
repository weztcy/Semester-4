```python
import pandas as pd
```

```python
kip = pd.read_csv('dataset.csv')
```

```python
data = kip.copy()
```

```python
target = 'penerima_kip'
```

```python
encode = ['peserta_kps', 'alat_transportasi', 'rekomendasi_bsm']
```

```python
data
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    dummy = dummy.astype(int)
    data = pd.concat([data, dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target_mapper = {'Tidak':0,'Ya':1}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['penerima_kip'] = data['penerima_kip'].apply(target_encode)
```

```python
data
```

```python
x = data.drop('penerima_kip', axis=1)
y = data['penerima_kip']
```

```python
x
```

```python
y
```

```python
# Model NBC
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x,y)
```

```python
# Menyimpan Model NBC
import pickle
pickle.dump(model, open('modelNBC_kip.pkl', 'wb'))
```
